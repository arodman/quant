```python
#Finally, Blitz and de Groot (2014) report
#a low-volatility effect in commodities. Similarly, we sort commodities
#on the standard deviation of returns in the past 36 months. The
#volatility factor is long (short) for commodities with the lowest
#(highest) return standard deviation.


#36 month trailing stsdanrd deisioant of returns. we go long 


#and finally a coefficient of variation factor defined
#as the standard deviation of returns between months t-36 to t-1
#you divided by the average return over this
#36*22=796


#Volatility We rank commodities on the 36-month trailing standard deviation of returns. We go long (short) in 30% of the
#commodities with the lowest (highest) volatility.

d015_vol0=data001
d015_vol0.rename(columns={'date':'Date'}, inplace=True)
d015_vol0 = d015_vol0.set_index('Date')
window_v=36*22


list_loop=list_x0
for x in range(len(list_loop)): 
    print(x)
    variable=str(list_loop[x])  
    print(variable)
    d015_vol0[str(variable+"_v30d_std36m")]=d015_vol0[str("v30d_"+variable)].rolling(window=window_v).std() 



list_xvo0 = [item if i%1 else item + "_v30d_std36m" for i,item in enumerate(list_x0)]


d015_vol1 = pd.DataFrame(data=d015_vol0, columns=list_xvo0)
d015_vol1 = d015_vol1.tail(1700)
#d010_cvar1.to_clipboard()
d015_vol2 = d015_vol1.rank(axis=1, method='dense',ascending =True).astype(int)  
d015_vol2['max_value'] = d015_vol2.max(axis=1)
r=6
d015_vol2['limit_long'] = r
d015_vol2['limit_short'] = d015_vol2['max_value'] - r
print(d015_vol2)
d015_vol3=d015_vol2

###19 -1
###1 1
#c_limit_down = 3
#c_limit_up = 17
c_signal_short = -1
c_signal_long = 1

def vol_f(row):
    if row['value'] <= row['limit_long']:
        val = c_signal_long
    elif row['value'] >= row['limit_short']:
        val = c_signal_short
    else:
        val = 0
    return val

list_loop=list_xvo0
for x in range(len(list_loop)):
    print(x)
    variable=str(list_loop[x])  
    #variable=str(variable)+"_r"
    variable=str(variable)
    print(variable)
    d015_vol3['value']=d015_vol3[str(variable)]
    d015_vol3['signal_output'] = d015_vol3.apply(vol_f, axis=1)
    d015_vol3.rename(columns={"signal_output":(str(variable)+"_015vol_s")}, inplace=True)   


    
d015_vol4=pd.merge(d015_vol1, d015_vol3,  how='left', left_on=['Date'], right_on = ['Date'])
#pyhton cftc4.to_clipboard()


d015_vol4.to_clipboard()    
```
