```python
G10
Oil
Equities
Relative yield curve
Rate diff
Relative equities
Ww change spot



EM
Comodity
World equities spx
Rate differential
Lcoail equities
Local CDS
Sport

STEER™: Short-term fair value model that captures the contemporaneous relationships between FX and other financial markets (rates, equity, commodities).
STEER™ is based on cointegration theory, meaning that leading and lagging relationships between assets are considered in the estimation but are not explicitly defined or observable by the model.
Excel
.. For altam, 
+10bps
Auction

--------------
-----------


GLOBAL EQUITIES
RELATIVE EQUITIES
REALTIVE YIELD CURVE
OIL 
RATES DIFF

10 
Oil
Equities
Relaitve yiedl cuirve
Rate diff
Relative equities
Cahnge in sport




EURUSD
GBPUSD
EURNOK
EURSEK
AUDUSD
NZDUSD
USDCAD
USDJPY
EURCHF



JPY




#### mxwo Comdty
##MXWO iNDEX
###eursek curncy

                                       'BBGTKR::BLOOMBERG::GTNOK10Y GOVT',
                                       'BBGTKR::BLOOMBERG::GTNOK2Y GOVT',
        
        
GTCHF3M GOVT
GTDEM3M GOVT
GTNOK10Y GOVT
GTNOK2Y GOVT
GTNZD10Y GOVT
GTNZD2Y GOVT
GTSEK10Y GOVT
GTSEK2Y GOVT
GTCAD10Y GOVT
GTCAD2Y GOVT

MXWO INDEX
EURSEK CURNCY     

#Interest rate differential	2Y rate	
#Relative yield curve steepness	10Y rate	10Y - 3m spread
#	3 month rate	
#CDS (for EM)	CDS	CDS for EM instead of YC steepness
#Local EQ	EQ index	Relative performance for G10 and absolute performance for EM
#Global EQ	MSCI ACWI	
#Commodity prices		for EM, based on tailored commodity baskets
#Oil price	WTI	for DM

#:21:28 no we do contemporaneous. so relative yield curve today would explain fv today
oil LOG(CL1 )
World Equities: log (msci mxaw)
Relative equities:   log msci country 1 - log msci country 2
Rates difference: log 2y rate country 1 - log 2y rate country 2
Relative Yield curve: (3m-10y of country 1) - (3m-10y of country 2) 
    


   A caveat: Raw prices or log prices?
Should we use raw prices or log prices in the cointegration tests? One key theme in this
article is that prices are I(1) series (which includes log-prices), and returns are I(0) series.
I(0) series can be obtained by differencing the I(1) series. So it comes naturally that log prices
fit this description better, for the difference of log prices is directly log returns, but the difference 
of raw prices is not percentage returns yet. However, according to (Alexander, 2002), “Since it is normally
the case that log prices will be cointegrated when the actual prices are cointegrated, it is standard, 
but not necessary, to perform the cointegration analysis on log prices.” So it is OK to analyze raw prices,
but log prices are preferable.

```

```python
from statsmodels.tsa.stattools import adfuller
def adfuller_test(series, signif=0.05, name='', verbose=False):
    """Perform ADFuller to test for Stationarity of given series and print report"""
    r = adfuller(series, autolag='AIC')
    output = {'test_statistic':round(r[0], 4), 'pvalue':round(r[1], 4), 'n_lags':round(r[2], 4), 'n_obs':r[3]}
    p_value = output['pvalue'] 
    def adjust(val, length= 6): return str(val).ljust(length)

    # Print Summary
    print(f'    Augmented Dickey-Fuller Test on "{name}"', "\n   ", '-'*47)
    print(f' Null Hypothesis: Data has unit root. Non-Stationary.')
    print(f' Significance Level    = {signif}')
    print(f' Test Statistic        = {output["test_statistic"]}')
    print(f' No. Lags Chosen       = {output["n_lags"]}')

    for key,val in r[4].items():
        print(f' Critical value {adjust(key)} = {round(val, 3)}')

    if p_value <= signif:
        print(f" => P-Value = {p_value}. Rejecting Null Hypothesis.")
        print(f" => Series is Stationary.")
    else:
        print(f" => P-Value = {p_value}. Weak evidence to reject the Null Hypothesis.")
        print(f" => Series is Non-Stationary.")
        
# ADF Test on each column
for name, column in X.iteritems():
    print(column.name)
    print(column.name)
    adfuller_test(column, name=column.name)
    print()
```

```python
from statsmodels.tsa.stattools import kpss
def kpss_test(x, h0_type='c'):
    indices = ['Test Statistic', 'p-value', '# of Lags']
    kpss_test = kpss(x, regression=h0_type, nlags ='auto')
    results = pd.Series(kpss_test[0:3], index=indices)
    for key, value in kpss_test[3].items():
        results[f'Critical Value ({key})'] = value
        return results
print('KPSS-EURUSD:')
print(kpss_test(X.eurusd))
print('___________________')
print('KPSS-GBPUSD:')
print(kpss_test(X.gbpusd))
print('___________________')
print('KPSS-USDJPY:')
print(kpss_test(X.usdjpy))
print('___________________')
print('KPSS-GC:')
print(kpss_test(X.gc))
print('___________________')
print('KPSS-NQ:')
print(kpss_test(X.nq))
print('___________________')
print('KPSS-ES:')
print(kpss_test(X.es))
```

```python
# Granger causality test

from statsmodels.tsa.stattools import grangercausalitytests
max_lag = 6
test = 'ssr_chi2test'
def causation_matrix(data, variables, test='ssr_chi2test', verbose=False):
    X = DataFrame(np.zeros((len(variables), len(variables))), columns=variables, index=variables)
    for c in X.columns:
        for r in X.index:
            test_result = grangercausalitytests(data[[r, c]], maxlag = max_lag, verbose = False)
            p_values = [round(test_result[i+1][0][test][1],4) for i in range(max_lag)]
            if verbose: print(f'Y = {r}, X = {c}, P Values = {p_values}')
            min_p_value = np.min(p_values)
            X.loc[r, c] = min_p_value
    X.columns = [var + '-x axis' for var in variables]
    X.index = [var + '-y axis' for var in variables]
    return X
causation_matrix(X, variables = X.columns)
```
