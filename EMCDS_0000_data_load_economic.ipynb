```python
#defien fucitons 

def get_frequency(data, reverse=False):
    ''' Compute the frequency of the underlying data '''
    if reverse:
        i = 1
    else:
        i = -1
    data_freq = (data.index[i].month - data.index[2*i].month)%12
    if data_freq==0:
        data_freq = (data.index[i].year - data.index[2*i].year)*12
    return data_freq

def compute_sa(df):
    df = df.fillna(method='ffill').dropna()
    freq = get_frequency(df)
    lookback = int(12 / freq)
    # longrun_mean = df.head(lookback * 5).mean()
    longrun_mean = df.loc[:dt.date(2018, 1, 1),:].mean()
    # df_res = (df/longrun_mean - 1).head(lookback * 5)
    df_res = (df / longrun_mean - 1).loc[:dt.date(2018, 1, 1),:]
    df_sa = df.copy()
    for i in range(0, lookback):
        df_sa.iloc[i::lookback, :] = df.iloc[i::lookback, :]*(1-df_res.iloc[i::lookback, :].mean())
    x = df_sa
    return x
```

```python
for x in range(len(c)):      
    #model=str(lst[x]) 
    print(x)
    x0=(x+1)
    country=[str(c[x])]
    country_str=str(c[x])
    print(country)
    country = pd.DataFrame(country,columns=["country"])
    print(country)
    search= pd.merge(country, m3_wb , how='left', left_on=["country"], right_on = ['country']) 
    print(search)
    for y in range(len(ind)):  #country=str(ind[x])
        print(y)
        y0=(y+1)
        print(y0)
        indicator=str(ind[y])
        print(indicator)
        search1=search[search['variable']==indicator]
        print(search1)
        symbols=search1.ticker.tolist()
        print(symbols)
        wb=reader.get_daily_ticks(symbols=symbols,start=start,end=end,fields=['PRICE'])
        wb.reset_index(inplace=True)
        wb["country"]=str(country_str)
        wb.columns=['time',str(indicator),'country']
        wb['year']=wb['time'].astype(str).str[:4]
        wb=wb[['year','country',str(indicator)]]
        print(wb) 
        if y0 == 1:
            ####if y==1, then 
            wb1=wb
            ###if y >1 thne
        else:
            wb1= pd.merge(wb1, wb , how='left', left_on=['year','country'], right_on = ['year','country']) 
            print(wb1)                      
    if x0 == 1:
    ####if y==1, then 
        wb2=wb1
        print(wb2)
    ###if y >1 thne
    else:
        wb2=pd.concat([wb2, wb1], ignore_index=True)
        print(wb2)
    
    
print(wb2)   
```
