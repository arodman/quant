```python
# convert series to supervised learning
def series_to_supervised(data, n_in=1, n_out=1, dropnan=True):
	n_vars = 1 if type(data) is list else data.shape[1]
	df = DataFrame(data)
	cols, names = list(), list()
	# input sequence (t-n, ... t-1)
	for i in range(n_in, 0, -1):
		cols.append(df.shift(i))
		names += [('var%d(t-%d)' % (j+1, i)) for j in range(n_vars)]
	# forecast sequence (t, t+1, ... t+n)
	for i in range(0, n_out):
		cols.append(df.shift(-i))
		if i == 0:
			names += [('var%d(t)' % (j+1)) for j in range(n_vars)]
		else:
			names += [('var%d(t+%d)' % (j+1, i)) for j in range(n_vars)]
	# put it all together
	agg = pd.concat(cols, axis=1)
	agg.columns = names
	# drop rows with NaN values
	if dropnan:
		agg.dropna(inplace=True)
	return agg
```

```python
dataset = rnn_brazil
country_iteration = "brazil"
dataset=dataset.set_index('date')
#print(dataset)
#len(dataset)
values = dataset.values
# ensure all data is float
values = values.astype('float32')
# normalize features
scaler = MinMaxScaler(feature_range=(0, 1))
scaled = scaler.fit_transform(values)


# specify the number of lag hours
n_hours = 1
n_features = 16
# frame as supervised learning
reframed = series_to_supervised(scaled, n_hours, 1)
print(reframed.shape)

hidden=10
epochs=50
batch_size=72
#n_train_hours=len(values)*0.75
#n_train_hours=1400
n_train_hours=1000
#n_train_hours = 365 * 24
n_obs = n_hours * n_features
values = reframed.values



#JUEGO DE HIDEN
#JUEFOG DE RPOGRESSION,
#JIEGOP DE OPTIMIZAICONT
#'GRAFICOS DE ERROR
#JUEGOS DESDE DODNE EMPIZA ELA VALIDATAION TRAIN Y TEST'
title_text0='RNN Prediction '

rnn_list1 = [1000,1200,1400]
rnn_list2 = [10,20,30]
#del history
#del history.history

epochs=50
batch_size=72
verbose=2

x_label= "Time"
y_label = "5Y CDS"
#del history.history['val_loss']

parameter_1=7
parameter_2=5
parameter_3=12
parameter_4=5
size_title=13
```

```python
os.chdir('K:/2020_2431/q/cds_report/images/rnn')
print(os.getcwd())

aggregated_n =0

for x in range(len(rnn_list1)):      
    n_train_hours=(rnn_list1[x])  
    print(n_train_hours)
    
    for y in range(len(rnn_list2)):  
        hidden=(rnn_list2[y]) 
        print(hidden)   
        
        
        ####### LSTM
        ###PART1
        # split into train and test sets
        l_train = values[:n_train_hours, :]
        l_test = values[n_train_hours:, :]

        # split into input and outputs
        l_train_X, l_train_y = l_train[:, :n_obs], l_train[:, -n_features]
        l_test_X, l_test_y = l_test[:, :n_obs], l_test[:, -n_features]
        print(l_train_X.shape, len(l_train_X), l_train_y.shape)

        # reshape input to be 3D [samples, timesteps, features]
        l_train_X = l_train_X.reshape((l_train_X.shape[0], n_hours, n_features))
        l_test_X = l_test_X.reshape((l_test_X.shape[0], n_hours, n_features))
        print(l_train_X.shape, l_train_y.shape, l_test_X.shape, l_test_y.shape)

        # design network
        l_model = Sequential()
        l_model.add(LSTM(hidden, input_shape=(l_train_X.shape[1], l_train_X.shape[2])))
        l_model.add(Dense(1))
        l_model.compile(loss='mae', optimizer='adam')
        # fit network
        history = l_model.fit(l_train_X, l_train_y, epochs=epochs, batch_size=72, validation_data=(l_test_X, l_test_y), verbose=2, shuffle=False)
        # plot history       
        
        ###PART2
        # make a prediction
        l_yhat = l_model.predict(l_test_X)
        print(l_yhat.shape)
        l_test_X = l_test_X.reshape((l_test_X.shape[0], n_hours*n_features))
        print(l_test_X.shape)
        # invert scaling for forecast
        print(l_test_X)
        l_inv_yhat = concatenate((l_yhat, l_test_X[:, -(n_features-1):]), axis=1)
        print(l_inv_yhat.shape)
        l_inv_yhat = scaler.inverse_transform(l_inv_yhat)
        l_inv_yhat = l_inv_yhat[:,0]

        # invert scaling for actual
        l_test_y = l_test_y.reshape((len(l_test_y), 1))
        l_inv_y = concatenate((l_test_y, l_test_X[:, -(n_features-1):]), axis=1)
        l_inv_y = scaler.inverse_transform(l_inv_y)
        l_inv_y = l_inv_y[:,0]
        # calculate RMSE
        #l_rmse = sqrt(mean_squared_error(l_inv_y, l_inv_yhat))
        #l_rmse = round(l_rmse,3)
        #l_rmse_text= " Test RMSE: %.3f"+str(l_rmse)
        #print('Test RMSE: %.3f' % l_rmse)
        
        
        ###PART3
        #tile for LSTM Models
        Met = "LSTM"
        title_text_l=title_text0+ country_iteration+" Met "+Met+" Train periods "+str(n_train_hours) +" Hidden Layers " +str(hidden)        
        
        name_l="Train and Test loss" +" Met "+ Met+" Train periods "+str(n_train_hours) +" Hidden Layers " +str(hidden) 
        fig,ax = plt.subplots(figsize=(parameter_1,parameter_2))
        plt.plot(history.history['loss'], label='train')
        plt.plot(history.history['val_loss'], label='test')
        plt.title(str(name_l), size=size_title)
        plt.legend()
        plt.savefig(str(name_l+".jpg"), bbox_inches='tight')
               
        name_l_p="Plot" +" Met "+ Met+" Train periods "+str(n_train_hours) +" Hidden Layers " +str(hidden) 
        #Plotting the Results
        fig,ax = plt.subplots(figsize=(parameter_3,parameter_4))
        plt.plot(l_inv_y, color = 'black', label = '5Y CDS Price')
        plt.plot(l_inv_yhat, color = 'blue', label = 'Predicted 5Y CDS Price')
        plt.title(str(title_text_l), size=size_title)
        plt.ylabel(str(y_label), size=20)
        plt.xlabel(str(x_label), size=20)
        plt.grid()
        plt.legend()
        plt.savefig(str(name_l_p+".jpg"), bbox_inches='tight')
        
        del history
        del l_model
        del l_train
        del l_test
        del l_inv_y
        
        del title_text_l
        del name_l
        del name_l_p
        
        
        
        ####### GRU
        ###PART1
        # split into train and test sets
        g_train = values[:n_train_hours, :]
        g_test = values[n_train_hours:, :]

        # split into input and outputs
        g_train_X, g_train_y = g_train[:, :n_obs], g_train[:, -n_features]
        g_test_X, g_test_y = g_test[:, :n_obs], g_test[:, -n_features]
        print(g_train_X.shape, len(g_train_X), g_train_y.shape)

        # reshape input to be 3D [samples, timesteps, features]
        g_train_X = g_train_X.reshape((g_train_X.shape[0], n_hours, n_features))
        g_test_X = g_test_X.reshape((g_test_X.shape[0], n_hours, n_features))
        print(g_train_X.shape, g_train_y.shape, g_test_X.shape, g_test_y.shape)

        # design network
        g_model = Sequential()
        #return_sequences=True,
        g_model.add(GRU(hidden,  input_shape=(g_train_X.shape[1], g_train_X.shape[2])))
        g_model.add(Dense(1))
        g_model.compile(loss='mae', optimizer='adam')
        # fit network
        history = g_model.fit(g_train_X, g_train_y, epochs=epochs, batch_size=72, validation_data=(g_test_X, g_test_y), verbose=2, shuffle=False)
        # plot history

        ###PART2
        # make a prediction
        g_yhat = g_model.predict(g_test_X)
        g_test_X = g_test_X.reshape((g_test_X.shape[0], n_hours*n_features))
        # invert scaling for forecast
        g_inv_yhat = concatenate((g_yhat, g_test_X[:, -(n_features-1):]), axis=1)
        g_inv_yhat = scaler.inverse_transform(g_inv_yhat)
        g_inv_yhat = g_inv_yhat[:,0]

        # invert scaling for actual
        g_test_y = g_test_y.reshape((len(g_test_y), 1))
        g_inv_y = concatenate((g_test_y, g_test_X[:, -(n_features-1):]), axis=1)
        g_inv_y = scaler.inverse_transform(g_inv_y)
        g_inv_y = g_inv_y[:,0]
        # calculate RMSE
        #g_rmse = sqrt(mean_squared_error(g_inv_y, g_inv_yhat))
        # g_rmse = round(g_rmse,3)
        #g_rmse_text= " Test RMSE: %.3f"+str(g_rmse)


        
        
        ###PART3
        #tile for GRU Models
        Met = "GRU"
        title_text_l=title_text0+country_iteration + " Met "+ Met+" Train periods "+str(n_train_hours) +" Hidden Layers " +str(hidden)        
        
        name_l="Train and Test loss" +" Met "+ Met +" Train periods "+str(n_train_hours) +" Hidden Layers " +str(hidden) 
        fig,ax = plt.subplots(figsize=(parameter_1,parameter_2))
        plt.plot(history.history['loss'], label='train')
        plt.plot(history.history['val_loss'], label='test')
        plt.title(str(name_l), size=size_title)
        plt.legend()
        plt.savefig(str(name_l+".jpg"), bbox_inches='tight')
        
        
        name_l_p="Plot" +" Met "+ Met+" Train periods "+str(n_train_hours) +" Hidden Layers " +str(hidden) 
        #Plotting the Results
        fig,ax = plt.subplots(figsize=(parameter_3,parameter_4))
        plt.plot(g_inv_y, color = 'black', label = '5Y CDS Price')
        plt.plot(g_inv_yhat, color = 'green', label = 'Predicted 5Y CDS Price')
        plt.title(str(title_text_l), size=size_title)
        plt.ylabel(str(y_label), size=20)
        plt.xlabel(str(x_label), size=20)
        plt.grid()
        plt.legend()
        plt.savefig(str(name_l_p+".jpg"), bbox_inches='tight')
        
        
        
    

        
        ##Aggregated
        aggregated_n = aggregated_n+1
        
        title_text_la=title_text0+country_iteration+" Train periods "+str(n_train_hours) +" Hidden Layers " +str(hidden)           
        name_l_p="Plot" +" Train periods "+str(n_train_hours) +" Hidden Layers " +str(hidden) 
        #Plotting the Results
        fig,ax = plt.subplots(figsize=(parameter_3,parameter_4))
        plt.plot(g_inv_y, color = 'black',  linewidth=2,label = '5Y CDS Price')
        plt.plot(g_inv_yhat, color = 'green', linewidth=1,linestyle = 'dashed',label = 'LSTM Predicted 5Y CDS Price')
        plt.plot(l_inv_yhat, color = 'blue',  linewidth=1,linestyle = 'dashed',label = 'GRU Predicted 5Y CDS Price')
        plt.title(str(title_text_la), size=size_title)
        plt.ylabel(str(y_label), size=20)
        plt.xlabel(str(x_label), size=20)
        plt.grid()
        plt.legend()
        plt.savefig(str(name_l_p+".jpg"), bbox_inches='tight')
        plt.savefig(str("fig_"+str(aggregated_n)+".jpg"), bbox_inches='tight')
        
        del history
        del g_model
        del g_train
        del g_test
        del g_inv_y
        del g_inv_yhat
        del title_text_l
        del name_l
        del name_l_p
        
        del l_inv_yhat

        
        
        
#del aggregated_n
        
        
    
  
    
print("Complete")    
```
