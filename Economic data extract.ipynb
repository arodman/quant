```python
import pandas_datareader as pdr
import datetime
import pandas as pd
# Define the start and end dates
start_date = datetime.datetime(2000, 1, 1)
end_date = datetime.datetime(2024, 1, 1)


# Specify FRED codes for the desired indicators
codes = [
    "RSXFS",  # Retail Sales
    "INDPRO",  # Industrial Production
    "CPIAUCNS",  # Monthly Headline Inflation
    "CPILFENS",  # Monthly Core Inflation
    "PCEPI",  # Monthly PCE Deflator   
    "PCEPILFE",  # Monthly PCE Core Deflator
    "UMCSENT",  # Michigan Consumer Sentiment
    "CFNAIMA3",  #Chicago Fed National Activity Index (CFNAI MA3) ]


    "UNRATE",  # Unemployment Rate
    "CIVPART",  # Civilian Labor Force Participation Rate
    "PAYEMS",  # Nonfarm Payrolls
    "USPRIV",  # Private Nonfarm Payrolls
        "PCEC96",  # PCE Core Services Excluding Housing
    "MEDCPIM158SFRBCLE",  # FRB Cleveland Median PCE
    #"PCEPILMCTPIFRBCLE",  # FRB Dallas Trimmed Mean PCE
    #"M12MTVH22999USM557NNBR",  # FRB Atlanta Sticky Core
]
# Fetch data from FRED
data = pdr.get_data_fred(codes, start_date, end_date)

    
    
# Rename columns
data = data.rename(columns={
    'RSXFS': 'retail_sales',
    'INDPRO': 'industrial_prod',
    'CPIAUCNS': 'headline_cpi',
    'CPILFENS': 'core_cpi',
    'PCEPI': 'pce_deflator',
    'PCEPILFE': 'core_pce_deflator',
    'UMCSENT': 'michigan_consumer',
    'CBSCS': 'Conference Board Consumer Sentiment',
    'CFNAIMA3': 'chicago_activity',
    'ISMMAN': 'ISM Manufacturing Index',
    'ISMNON': 'ISM Services Index',
    'CPIAUCNSx1y': 'Michigan 1-Year Ahead Inflation',
    'CPIAUCNSx5y': 'Michigan 5-Year Ahead Inflation',
    'ICSA': 'Initial Jobless Claims',
    'CCSA': 'Continuing Jobless Claims',
    'UNRATE': 'ur',
    'CIVPART': 'Labor Force Participation Rate',
    'PAYEMS': 'Nonfarm Payrolls',
    'USPRIV': 'Private Nonfarm Payrolls',
    'AHE': 'Average Hourly Earnings',
    'CES0500000003': 'Average Weekly Earnings',
    'PCEC96':  'PCE Core Services Excluding Housing',
    'MEDCPIM158SFRBCLE': 'FRB Cleveland Median PCE'
})


# Display the data
print(data.head())

data00=data.copy()
#year on year
#3month annoualies
#6 monts a nnlauez
#mon on onth  
#core_pce_exh					
#retail_sales	industrial_prod	headline_cpi	core_cpi	pce_deflator	core_pce_deflator

data['nfp_diff']=data['Nonfarm Payrolls'].diff()
data['private_nfp_diff']=data['Private Nonfarm Payrolls'].diff()


usi_m_tickers=['nfp_diff','private_nfp_diff',
               'ur','michigan_consumer','chicago_activity',
               'industrial_prod','retail_sales']


hv_tb_us_000_001_m=data.copy()
#print(hv_tb_eu_000_001_m)
for x in range(len(usi_m_tickers)):
    print(x)
    variable=str(usi_m_tickers [x])
    variable0=variable +"_zscore5y"
    hv_tb_us_000_001_m[(str(variable)+"_rm")] = hv_tb_us_000_001_m[variable].rolling(window=12 * 5).mean()  # 12 months per year
    hv_tb_us_000_001_m[(str(variable)+"_sd")] = hv_tb_us_000_001_m[variable].rolling(window=12 * 5).std()
    # Calculate rolling z-score
    hv_tb_us_000_001_m[variable0] = (hv_tb_us_000_001_m[variable]- hv_tb_us_000_001_m[(str(variable)+"_rm")]) / hv_tb_us_000_001_m[(str(variable)+"_sd")]
    del variable0




#mcihignc a adn dnfp doifernetec 
#z scores moviong averages  
```
