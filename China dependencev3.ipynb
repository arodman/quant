```python
base_t0= pd.read_csv (r'K:\2020_2431\q\sandbox_china_comm\base_t0.csv')
base_t1= pd.read_csv (r'K:\2020_2431\q\sandbox_china_comm\base_t1.csv')
base_t2= pd.read_csv (r'K:\2020_2431\q\sandbox_china_comm\base_t2.csv')
base_t3= pd.read_csv (r'K:\2020_2431\q\sandbox_china_comm\base_t3.csv')
base_t4= pd.read_csv (r'K:\2020_2431\q\sandbox_china_comm\base_t4.csv')
base_t5= pd.read_csv (r'K:\2020_2431\q\sandbox_china_comm\base_t5.csv')

#print(base_t0)
print(base_t0.columns)

#print(base_t1)
print(base_t1.columns)

#print(base_t2)
print(base_t2.columns)

#print(base_t3)
print(base_t3.columns)


#print(base_t4)
print(base_t4.columns)


base_t0.rename(columns={'Trade Value (US$)':'reporter_comm_i_exp_world'}, inplace=True)
base_t1.rename(columns={'Trade Value (US$)':'reporter_exp_world'}, inplace=True)
base_t2.rename(columns={'Trade Value (US$)':'reporter_comm_i_exp_china'}, inplace=True)
base_t3.rename(columns={'Trade Value (US$)':'china_imp_comm_i'}, inplace=True)
base_t4.rename(columns={'Trade Value (US$)':'world_imp_comm_i'}, inplace=True)
base_t5.rename(columns={'Trade Value (US$)':'world_exp_comm_i'}, inplace=True)

base_t0=base_t0[["Year", "Trade Flow", "Reporter Code", "Reporter","Commodity Code", "Commodity","reporter_comm_i_exp_world"]]
data=base_t0

base_t1=base_t1[["Year","Reporter","reporter_exp_world"]]
data=pd.merge(data, base_t1,  how='left', left_on=["Year","Reporter"], right_on = ["Year","Reporter"])

base_t2=base_t2[["Year","Reporter","Commodity Code","reporter_comm_i_exp_china"]]
data=pd.merge(data, base_t2,  how='left', left_on=["Year","Reporter","Commodity Code"], right_on = ["Year","Reporter","Commodity Code"])

base_t3=base_t3[["Year","Commodity Code","china_imp_comm_i"]]
data=pd.merge(data, base_t3,  how='left', left_on=["Year","Commodity Code"], right_on = ["Year","Commodity Code"])

base_t4=base_t4[["Year","Commodity Code","world_imp_comm_i"]]
data=pd.merge(data, base_t4,  how='left', left_on=["Year","Commodity Code"], right_on = ["Year","Commodity Code"])

base_t5=base_t5[["Year","Commodity Code","world_exp_comm_i"]]
data=pd.merge(data, base_t5,  how='left', left_on=["Year","Commodity Code"], right_on = ["Year","Commodity Code"])




data.to_csv(r'K:\2020_2431\q\sandbox_china_comm\base_t6.csv', index = True, header=True)
```

```python
#reporter_comm_i_exp_world	reporter_exp_world	reporter_comm_i_exp_china	china_imp_comm_i	world_imp_comm_i


data["Index_0"]= (data.reporter_comm_i_exp_world/data.reporter_exp_world)
data["Index_1"]= (data.reporter_comm_i_exp_china/data.reporter_comm_i_exp_world)
data["Index_2"]= (data.china_imp_comm_i/data.world_imp_comm_i)+(1-(data.reporter_comm_i_exp_world/data.world_exp_comm_i))/2
data["Index_DC"]=data["Index_0"]*data["Index_1"]*data["Index_2"]
data["Index_DC"]=data["Index_DC"]**(1./3.)


print(data)
data.to_clipboard()
```
