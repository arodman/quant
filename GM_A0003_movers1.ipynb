```python
###GENERATE THE DATA

#list_x0 = ['BRL','COP','MXN','CLP','PEN','CZK','HUF','PLN','RUB','ILS','ZAR','TRY','IDR','THB','MYR','KRW','PHP']
list_y0 = ["cds5y_","fx_","l_msci_","hccy_","lccy_"]
## inr,twd
list_x0 = ['BRL','CLP','CNH','COP','CZK','IDR','ILS','INR','KRW','MXN','PEN','PLN','RUB','HUF','ZAR','THB','TWD','MYR','TRY']
list_z0 = ['1_Week','1_Month','3_Months','6_Months']
list_z1 = ['5','22','66','132']
diff0=5
diff1=22
diff2=66
diff3=132
```

```python
###variatons 
#latest vlaues  - avera
#latest value - averar / standar decaitons

mm=pd.DataFrame()
mm["date"]=data_z["date"]

loop_list1=list_y0
loop_list2=list_x0
loop_list3=list_z0
loop_list4=list_z1

for x in range(len(loop_list1)): 
    print(x)
    variable=str(loop_list1[x])  
    print(variable)
    
    for y in range(len(loop_list2)): 
        print(y)
        country=str(loop_list2[y])  
        print(country)       
        c_v=str(variable+country)
        for z in range(len(loop_list3)):
            print(z)
            time_l=str(loop_list3[z])            
            print(time_l)
            window=str(loop_list4[z])
            print(window)
            window=int(window)+0                 
            try:
                mm[str(c_v)]=data_z[str(c_v)]
                mm[str("m_"+c_v)]=data_z[str(c_v)].rolling(window=window).mean()
                mm[str("s_"+c_v)]=data_z[str(c_v)].rolling(window=window).std() 
                mm[str("lg_"+time_l+"_"+c_v)]= (data_z[str(c_v)].shift((window)))
                mm[str("var_"+time_l+"_"+c_v)]=((mm[str(c_v)]/mm[str("lg_"+time_l+"_"+c_v)])-1)
                mm[str("z_score_"+time_l+"_"+c_v)]=((mm[str(c_v)]-mm[str("m_"+c_v)])/mm[str("s_"+c_v)])           
            except KeyError:
                pass
            
#########################################
#                                                      
mm=mm.fillna(method='ffill')
mm=mm.tail(1)            
print("Complete")
```
